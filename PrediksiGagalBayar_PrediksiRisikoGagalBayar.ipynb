# Projek : **Penerapan Machine Learning untuk Prediksi Risiko Kredit dan Estimasi Kerugian Menggunakan Algoritma XGBoost**

# **Business Understanding**

### Problem Statements

Industri keuangan menghadapi tantangan dalam mengelola risiko kredit, terutama dengan bertambahnya jumlah pengajuan pinjaman setiap tahun. Dalam proses menentukan pemberian kredit, bank dan lembaga keuangan sering menghadapi risiko pembayaran yang gagal, yang dapat memengaruhi kestabilan keuangan mereka. Selain itu, memperkirakan kerugian dari kegagalan pembayaran menjadi bagian penting dalam mengelola kredit dengan baik.
1. Bagaimana memprediksi risiko kegagalan pembayaran kredit berdasarkan data historis pelanggan?
2. Bagaimana memperkirakan tingkat kerugian yang mungkin terjadi akibat kegagalan pembayaran kredit?
3. Dapatkah pendekatan berbasis machine learning meningkatkan akurasi prediksi dibandingkan dengan metode tradisional?

### Goals

1. Membuat prediksi risiko kredit yang lebih akurat untuk mengurangi kerugian akibat gagal bayar.
2. Memberikan perkiraan kerugian yang membantu lembaga keuangan mengatur portofolio kredit dengan lebih baik.
3. Mempermudah proses pengambilan keputusan kredit dengan pendekatan berbasis data dengan memanfaatkan teknologi machine learning.

### Solution statements

 Membangun model prediksi risiko kredit dan estimasi tingkat kerugian menggunakan algoritma XGBoost (Extreme Gradient Boosting). Model ini akan menggunakan data pelanggan, seperti informasi pribadi, riwayat pembayaran, dan kondisi keuangan, untuk mengidentifikasi faktor yang memengaruhi risiko gagal bayar. Selain itu, algoritma ini akan digunakan untuk memperkirakan kerugian yang mungkin terjadi jika pembayaran gagal. Hasil dari model ini akan membantu lembaga keuangan dalam menentukan tingkat risiko setiap pelanggan, merencanakan langkah pencegahan, dan mengatur kredit dengan lebih baik

# **Modeling**

Pada fase modeling, dua model prediktif dikembangkan menggunakan algoritma XGBoost (Extreme Gradient Boosting). Model pertama bertujuan untuk memprediksi apakah seorang pemohon pinjaman akan mengalami kegagalan dalam melakukan pembayaran, sementara model kedua dirancang untuk memprediksi tingkat kerugian yang mungkin terjadi akibat kegagalan pembayaran. XGBoost adalah algoritma machine learning berbasis ensemble yang menggunakan kombinasi pohon keputusan sebagai elemen dasarnya dan pendekatan Gradient Boosting sebagai metode utamanya untuk mengoptimalkan hasil prediksi (Abdurrahman et al., 2022). Algoritma ini diterapkan pada masalah supervised learning dengan memanfaatkan sejumlah fitur dari data latih xi untuk memperkirakan nilai variabel target yi​ (Mendrofa et al., 2023).

Metode XGBoost dipilih karena keunggulannya dalam menangani data yang memiliki jumlah fitur yang besar, serta kemampuannya untuk bekerja secara efisien dengan data yang tidak seimbang. Selain itu, algoritma ini terkenal karena kecepatan komputasinya, penggunaan memori yang efisien, serta kemampuannya untuk mencegah overfitting melalui penerapan regularisasi. Beberapa penelitian sebelumnya menunjukkan keunggulan algoritma ini. Salah satunya adalah penelitian oleh J. M. A. S. Dachin (2023), yang membandingkan XGBoost dengan algoritma Random Forest pada klasifikasi keputusan kredit. Penelitian tersebut menunjukkan bahwa XGBoost memiliki akurasi yang lebih tinggi dibandingkan dengan model Random Forest. Pada data berjumlah 10.000 entri, XGBoost menghasilkan akurasi 1.0, sementara Random Forest hanya menghasilkan akurasi 0.998. Pada data berjumlah 100.000 entri, XGBoost tetap menghasilkan akurasi 1.0, sementara Random Forest mencapai akurasi 0.999. Hasil ini semakin memperkuat kecocokan XGBoost untuk menangani kasus prediksi yang kompleks, seperti prediksi gagal bayar dan estimasi kerugian dalam skenario peminjaman.

Namun, algoritma XGBoost juga memiliki beberapa keterbatasan. Salah satunya adalah waktu yang dibutuhkan untuk tuning parameter yang lebih lama dibandingkan algoritma lain, mengingat banyaknya hyperparameter yang perlu dioptimalkan untuk mencapai performa terbaik (Siringoringo et al., 2022). Selain itu, meskipun XGBoost dapat menangani data yang tidak seimbang, performanya bisa menurun pada data yang sangat tidak seimbang, meskipun hal ini dapat diatasi dengan menyesuaikan bobot kelas atau menggunakan teknik seperti oversampling atau undersampling.

# **Evaluation**

Untuk mengevaluasi model prediksi gagal bayar, kami menggunakan beberapa metrik evaluasi yang relevan dengan masalah klasifikasi biner, seperti AUC (Area Under the Curve), ROC (Receiver Operating Characteristic), dan KS (Kolmogorov-Smirnov).

1. ROC Curve
  - ROC Curve adalah grafik yang menunjukkan performa model klasifikasi biner dengan membandingkan True Positive Rate (TPR) dengan False Positive Rate (FPR) pada berbagai ambang batas (threshold). TPR menunjukkan proporsi kasus positif yang terdeteksi dengan benar, sementara FPR menunjukkan proporsi kasus negatif yang salah terdeteksi sebagai positif.
  - Dalam hal ini, nilai TPR menunjukkan proporsi pinjaman yang benar-benar gagal bayar (positif) yang berhasil diprediksi oleh model, sementara FPR menunjukkan proporsi pinjaman yang sebenarnya tidak gagal bayar (negatif) namun diprediksi sebagai gagal bayar oleh model.
  - Semakin jauh ROC Curve dari garis diagonal (yang menunjukkan prediksi acak), semakin baik model tersebut dalam klasifikasi biner.

2. AUC (Area Under the Curve)
  - Untuk mengukur luas area di bawah kurva ROC.
  - Nilai AUC berkisar antara 0 hingga 1, dengan nilai yang lebih tinggi menunjukkan model yang lebih baik. AUC yang mendekati 1 menunjukkan model yang sangat baik dalam memisahkan kasus positif dan negatif, sedangkan AUC yang lebih rendah menunjukkan bahwa model memiliki kemampuan yang lebih buruk.

3. KS (Kolmogorov-Smirnov) Statistic
  - Untuk mengukur sejauh mana distribusi kumulatif prediksi untuk kasus positif (gagal bayar) dan negatif (tidak gagal bayar) terpisah.
  - KS dihitung dengan mencari selisih maksimum antara distribusi kumulatif untuk kasus positif dan negatif.
  - Dalam hal ini, jika nilai KS tinggi, itu menunjukkan bahwa model mampu dengan baik memisahkan kedua kelas ini pada berbagai ambang batas probabilitas, dengan distribusi kumulatif kasus gagal bayar terpisah jauh dari distribusi kasus tidak gagal bayar. Sebaliknya, nilai KS yang rendah menunjukkan bahwa model kesulitan dalam memisahkan kasus positif dan negatif.

Untuk evaluasi model prediksi kerugian gagal bayar, kami menggunakan beberapa metrik evaluasi yang umum digunakan dalam regresi untuk menilai seberapa baik model dalam memprediksi nilai kerugian. Metrik yang digunakan dalam proyek ini adalah Mean Squared Error (MSE), Mean Absolute Error (MAE), dan Root Mean Squared Error (RMSE).
  1. Mean Squared Error (MSE)
    - MSE mengukur rata-rata dari kuadrat selisih antara nilai prediksi dan nilai aktual.
    - MSE lebih sensitif terhadap kesalahan besar karena kuadrat dari perbedaan.
    - Semakin kecil nilai MSE, semakin baik model dalam memprediksi nilai kerugian. Nilai MSE yang tinggi menunjukkan bahwa model cenderung membuat prediksi yang jauh dari nilai sebenarnya.
  2. Mean Absolute Error (MAE)
    - MAE mengukur rata-rata selisih absolut antara nilai prediksi dan nilai aktual.
    - MAE memberikan gambaran yang lebih sederhana dan lebih intuitif tentang rata-rata kesalahan prediksi, karena tidak memberikan penalti yang lebih besar untuk kesalahan besar seperti MSE.
    - MAE yang lebih kecil menunjukkan model yang lebih akurat dalam memprediksi kerugian.
  3. Root Mean Squared Error (RMSE)
    - RMSE adalah akar kuadrat dari MSE
    - RMSE mengukur seberapa besar kesalahan rata-rata yang terjadi pada prediksi model, dalam satuan yang sama dengan data asli.
    - RMSE memberikan penilaian kesalahan yang lebih terukur dan memberikan penalti lebih besar pada kesalahan besar. Semakin kecil nilai RMSE, semakin baik prediksi model.

### Evaluation Model 1

ROC Curve dan AUC

Di sini, kami menggunakan metode predict_proba() dari model XGBoost untuk mendapatkan probabilitas prediksi untuk setiap instance pada data uji (X_test). Probabilitas ini akan digunakan untuk menghitung True Positive Rate (TPR) dan False Positive Rate (FPR) pada berbagai ambang batas.

Setelah itu, data yang berisi nilai aktual dan probabilitas prediksi tersebut digabungkan dalam satu dataframe (df_actual_predicted_xgb). Data ini digunakan untuk menghitung FPR dan TPR menggunakan fungsi roc_curve() dari scikit-learn. Fungsi ini menghasilkan tiga variabel: FPR, TPR, dan threshold, yang diperlukan untuk membangun ROC Curve.

Kemudian, AUC (Area Under the Curve) dihitung dengan menggunakan roc_auc_score(), yang memberikan nilai yang menunjukkan seberapa baik model dalam membedakan antara kelas positif dan negatif.

```python
from sklearn.metrics import roc_curve, roc_auc_score

y_proba_xgb = xgb_model.predict_proba(X_test)[:, 1]

df_actual_predicted_xgb = pd.concat([pd.DataFrame(np.array(y_test), columns=['y_actual']), pd.DataFrame(y_proba_xgb, columns=['y_proba_xgb'])], axis=1)
df_actual_predicted_xgb.index = y_test.index

fpr, tpr, tr = roc_curve(df_actual_predicted_xgb['y_actual'], df_actual_predicted_xgb['y_proba_xgb'])
auc_xgb = roc_auc_score(y_test, y_proba_xgb)
print("AUC Score:", auc_xgb)
```

Dengan hasil AUC Score sebesar 0.9168, dapat diketahui bahwa model menunjukkan kinerja yang sangat baik dalam membedakan antara kelas positif dan negatif

Setelah menghitung nilai AUC, kami kemudian menggambar ROC Curve. Di dalam plot ini, FPR dipetakan pada sumbu x dan TPR pada sumbu y, sedangkan garis diagonal putus-putus menunjukkan model acak (nilai AUC = 0.5). Model yang lebih baik akan memiliki kurva yang lebih dekat dengan sudut kiri atas (di mana TPR tinggi dan FPR rendah), yang menunjukkan kemampuan model untuk mengidentifikasi kasus positif dengan lebih tepat.

```python
fpr, tpr, _ = roc_curve(y_test, y_proba_xgb)
plt.plot(fpr, tpr, label='AUC = %0.4f' %auc_xgb)
plt.plot(fpr, fpr, linestyle = '--', color='k')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
```

Berdasarkan luaran visualisasi yang diperoleh, dapat diketahui bahwa ROC curve menunjukkan sejauh mana model dapat membedakan antara kelas positif dan negatif. Semakin mendekati sudut kiri atas dari grafik, semakin baik model dalam mengklasifikasikan data dengan benar, dengan false positive rate (FPR) yang rendah dan true positive rate (TPR) yang tinggi. Grafik juga menunjukkan nilai AUC sebesar 0.9168, yang menunjukkan bahwa model memiliki kemampuan yang sangat baik dalam membedakan antara dua kelas.

KS Statistik

Selanjutnya, untuk menghitung KS (Kolmogorov-Smirnov) Statistic, kami memulai dengan mengurutkan data berdasarkan probabilitas prediksi untuk kasus gagal bayar (kolom y_proba_xgb). Proses ini dilakukan agar distribusi kumulatif kasus positif dan negatif bisa dibandingkan secara lebih efektif. Setelah itu, data direset indeksnya agar penomoran dimulai dari 0.

Setelah data terurut, kami menghitung beberapa kolom kumulatif:

- Cumulative N Population: Ini menunjukkan urutan baris yang dihitung berdasarkan indeks, yang mewakili jumlah total data yang ada.
- Cumulative N Bad: Ini adalah jumlah kumulatif dari kasus gagal bayar (bad cases) yang dihitung menggunakan cumsum() pada kolom y_actual.
- Cumulative N Good: Ini adalah jumlah kumulatif dari kasus tidak gagal bayar (good cases), yang dihitung dengan mengurangi jumlah kumulatif kasus buruk dari jumlah total data.
- Cumulative Perc Population: Persentase kumulatif populasi yang dihitung dengan membagi Cumulative N Population dengan jumlah total data.
- Cumulative Perc Bad: Persentase kumulatif dari kasus gagal bayar yang dihitung dengan membagi Cumulative N Bad dengan jumlah kasus gagal bayar dalam data.
- Cumulative Perc Good: Persentase kumulatif dari kasus tidak gagal bayar yang dihitung dengan membagi Cumulative N Good dengan jumlah kasus tidak gagal bayar dalam data.

```python
df_actual_predicted_xgb = df_actual_predicted_xgb.sort_values('y_proba_xgb')
df_actual_predicted_xgb = df_actual_predicted_xgb.reset_index()

df_actual_predicted_xgb['Cumulative N Population'] = df_actual_predicted_xgb.index + 1
df_actual_predicted_xgb['Cumulative N Bad'] = df_actual_predicted_xgb['y_actual'].cumsum()
df_actual_predicted_xgb['Cumulative N Good'] = df_actual_predicted_xgb['Cumulative N Population'] - df_actual_predicted_xgb['Cumulative N Bad']
df_actual_predicted_xgb['Cumulative Perc Population'] = df_actual_predicted_xgb['Cumulative N Population'] / df_actual_predicted_xgb.shape[0]
df_actual_predicted_xgb['Cumulative Perc Bad'] = df_actual_predicted_xgb['Cumulative N Bad'] / df_actual_predicted_xgb['y_actual'].sum()
df_actual_predicted_xgb['Cumulative Perc Good'] = df_actual_predicted_xgb['Cumulative N Good'] / (df_actual_predicted_xgb.shape[0] - df_actual_predicted_xgb['y_actual'].sum())
```

Data diatas berguna sebagai informasi untuk memahami seberapa efektif model dalam mengidentifikasi kasus gagal bayar sepanjang seluruh dataset. Berikut adalah penjelasan setiap kolom:

1. index: Ini adalah indeks atau nomor urut dari data yang sedang dianalisis. Pada baris pertama, indeksnya adalah 238535, yang merujuk pada pemohon pinjaman tertentu dalam data.

2. y_actual: Ini adalah nilai aktual (ground truth) dari variabel target untuk pemohon pinjaman ini. Nilai 0 menunjukkan bahwa pemohon ini tidak mengalami kegagalan pembayaran (yaitu, dia dianggap "good" atau "tidak gagal bayar").

3. y_proba_xgb: Ini adalah probabilitas prediksi dari model XGBoost untuk kasus ini. Nilai 0.000065 menunjukkan probabilitas bahwa pemohon pinjaman ini akan gagal bayar. Probabilitas ini dihitung oleh model berdasarkan berbagai fitur input yang telah dilatih sebelumnya.

4. Cumulative N Population: Ini menunjukkan jumlah kumulatif dari seluruh pemohon pinjaman yang telah dianalisis sampai titik ini. Pada baris pertama, hanya ada 1 pemohon yang dianalisis, karena baris ini adalah data pertama.

5. Cumulative N Bad: Ini adalah jumlah kumulatif kasus gagal bayar yang terdeteksi sampai titik ini. Pada baris pertama, tidak ada pemohon yang gagal bayar (karena y_actual = 0), sehingga jumlahnya adalah 0.

6. Cumulative N Good: Ini adalah jumlah kumulatif dari pemohon yang tidak gagal bayar. Pada baris pertama, hanya ada 1 pemohon yang tidak gagal bayar (karena y_actual = 0), sehingga jumlahnya adalah 1.

7. Cumulative Perc Population: Ini menunjukkan persentase kumulatif pemohon yang telah dianalisis dibandingkan dengan total jumlah data. Karena ini adalah baris pertama, persentasenya sangat kecil (0.000011), yang berarti hanya sebagian kecil dari data yang telah dipertimbangkan.

8. Cumulative Perc Bad: Ini menunjukkan persentase kumulatif dari kasus gagal bayar dibandingkan dengan total kasus gagal bayar dalam data. Pada baris pertama, tidak ada kasus gagal bayar, jadi nilai persentase ini adalah 0.0%.

9. Cumulative Perc Good: Ini menunjukkan persentase kumulatif dari kasus tidak gagal bayar dibandingkan dengan total kasus tidak gagal bayar dalam data. Karena pemohon ini tidak gagal bayar, persentase kumulatifnya sangat kecil (0.000012), menunjukkan bahwa pemohon yang "good" sangat sedikit dibandingkan dengan keseluruhan dataset pada tahap ini.

Secara keseluruhan, baris pertama ini memberikan gambaran tentang analisis kumulatif yang sedang dilakukan berdasarkan prediksi probabilitas model XGBoost untuk kasus pemohon pinjaman pertama.



Langkah terakhir adalah menghitung KS dengan mencari selisih maksimum antara Cumulative Perc Good dan Cumulative Perc Bad pada titik tertentu. Selisih ini menunjukkan seberapa terpisah distribusi antara kasus positif dan negatif pada probabilitas tertentu. Semakin besar nilai KS, semakin baik model dalam memisahkan kedua kelas.

Untuk memvisualisasikan hasil KS, kami menggambarkan dua kurva kumulatif: satu untuk kasus gagal bayar (berwarna merah) dan satu untuk kasus tidak gagal bayar (berwarna biru). Grafik ini memberikan gambaran mengenai seberapa baik model dalam membedakan antara dua kelas dengan memanfaatkan probabilitas prediksi yang dihasilkan.

```python
KS = max(df_actual_predicted_xgb['Cumulative Perc Good'] - df_actual_predicted_xgb['Cumulative Perc Bad'])

plt.plot(df_actual_predicted_xgb['y_proba_xgb'], df_actual_predicted_xgb['Cumulative Perc Bad'], color='r')
plt.plot(df_actual_predicted_xgb['y_proba_xgb'], df_actual_predicted_xgb['Cumulative Perc Good'], color='b')
plt.xlabel('Estimated Probability for Being Bad')
plt.ylabel('Cumulative %')
plt.title('Kolmogorov-Smirnov:  %0.4f' %KS)
```

Perbedaan antara kedua kurva ini pada berbagai tingkat probabilitas mencerminkan kemampuan model dalam memisahkan pemohon yang gagal bayar (bad) dengan yang tidak gagal bayar (good). Berdasarkan data di atas, dapat diketahui bahwa nilai KS sebesar 0.6588 menunjukkan bahwa model cukup efektif dalam memisahkan pemohon yang gagal bayar dan yang tidak.

Model yang dibangun menghasilkan performa AUC = 0.9168 dan KS = 0.6588. Pada dunia credit risk modeling, umumnya AUC di atas 0.7 dan KS di atas 0.3 sudah termasuk performa yang baik.
